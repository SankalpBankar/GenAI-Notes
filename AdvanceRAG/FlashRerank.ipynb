**Flash Rank**: Ultra-lite & Super-fast Python library for search & retrieval re-ranking.

1. Ultra-lite: No heavy dependencies. Runs on CPU with a tiny ~4MB reranking model.
2. Super-fast: Speed depends on the number of tokens in passages and query, plus model depth.
3. Cost-efficient: Ideal for serverless deployments with low memory and time requirements.
4. Based on State-of-the-Art Cross-encoders: Includes models like ms-marco-TinyBERT-L-2-v2 (default), ms-marco-MiniLM-L-12-v2, rank-T5-flan, and ms-marco-MultiBERT-L-12.
5. Sleek Models for Efficiency: Designed for minimal overhead in user-facing scenarios.

Flash Rank is tailored for scenarios requiring efficient and effective reranking, balancing performance with resource usage.

**Model Options**:

1. Nano: ~4MB, blazing fast model with competitive performance (ranking precision).
2. Small: ~34MB, slightly slower with the best performance (ranking precision).
3. Medium: ~110MB, slower model with the best zero-shot performance (ranking precision).
4. Large: ~150MB, slower model with competitive performance (ranking precision) for 100+ languages.

```python
from flashrank.Ranker import Ranker, RerankRequest
```

```python
def get_result(query, passages, choice):
  if choice == "Nano":
    ranker = Ranker()
  elif choice == "Small":
    ranker = Ranker(model_name="ms-marco-MiniLM-L-12-v2", cache_dir = "/opt")
  elif choice == "Medium":
    ranker = Ranker(model_name="rank-T5-flan", cache_dir = "/opt")
  elif choice == "Large":
    ranker = Ranker(model_name="ms-marco-MultiBERT-L-12", cache_dir="/opt")

  rerankrequest = RerankRequest(query=query, passages=passages)
  results = ranker.rerank(rerankrequest)
  print(results)
  return results
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
```

```python
documents = TextLoader("/content/mauritius_places.txt").load()
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap = 100)
```

```python
texts = text_splitter.split_documents(documents)
```

```python
for id, text in enumerate(texts):
  text.metadata["id"] = id
```

```python
embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
from langchain.retrievers import ContextualCompressionRetriever
from langchain.document_compressors import FlashrankRerank
from langchain_groq import ChatGroq
```

```python
llm = ChatGroq(temperature=0)
```

```python
compressor = FlashrankRerank()
```

```python
compression_retriever = ContextualCompressionRetriever(base_compressor=compressor, base_retriever=retriever)
```

```python
compressed_docs = compression_retriever.invoke("List names of hotels in Mauritius for stay.")
```

```python
len(compressed_docs)
```

```python
print([doc.metadata["id"] for doc in compressed_docs])
```

```python
pretty_print_docs(compressed_docs)
```

```python
from langchain.chains import RetrievalQA

chain = RetrievalQA.from_chain_type(llm=llm, retriever=compression_retriever)
```
